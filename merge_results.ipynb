```python
from os import listdir
from os.path import isfile, join

import numpy as np
import pandas as pd
```

```python
conv_logs_path = "/mnt/d/UFRGS/TCC/Dados/repos/sb_ref_kan/models/Conv/ConvLogs/"
energy_logs_path = "/mnt/d/UFRGS/TCC/Dados/repos/sb_ref_kan/models/Energy/EnergyLogs/"

conv_logs_files = [f for f in listdir(conv_logs_path) if isfile(join(conv_logs_path, f))]
energy_logs_files = [f for f in listdir(energy_logs_path) if isfile(join(energy_logs_path, f))]
```

```python
energy_logs_files
```

```python
conv_logs = pd.read_csv(conv_logs_path + conv_logs_files[0], sep=";")

conv_loss = conv_logs[["iteration", "loss"]]
conv_et = conv_logs[["iteration", "elapsed_time"]]

conv_loss.columns = ["iteration", conv_logs_files[0][:-4] + "_loss"]
conv_et.columns = ["iteration", conv_logs_files[0][:-4] + "_et"]

for f in conv_logs_files[1:]:
    print(f)
    model_name = f[:-4]
    df = pd.read_csv(conv_logs_path + f, sep=";")
    conv_loss[model_name + "_loss"] = df["loss"]
    conv_et[model_name + "_et"] = df["elapsed_time"]
```

```python
energy_logs = pd.read_csv(energy_logs_path + energy_logs_files[0], sep=";")

energy_loss = energy_logs[["iteration", "loss"]]
energy_et = energy_logs[["iteration", "elapsed_time"]]

energy_loss.columns = ["iteration", energy_logs_files[0][:-4] + "_loss"]
energy_et.columns = ["iteration", energy_logs_files[0][:-4] + "_et"]

for f in energy_logs_files[1:]:
    print(f)
    model_name = f[:-4]
    df = pd.read_csv(energy_logs_path + f, sep=";")
    energy_loss[model_name + "_loss"] = df["loss"]
    energy_et[model_name + "_et"] = df["elapsed_time"]
```

```python
energy_loss
```

```python
conv_loss.to_csv(conv_logs_path + "ConvLoss.csv")
conv_et.to_csv(conv_logs_path + "ConvEt.csv")

energy_loss.to_csv(energy_logs_path + "EnergyLoss.csv")
energy_et.to_csv(energy_logs_path + "EnergyEt.csv")
```
